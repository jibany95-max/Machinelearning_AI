```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import numpy as np
```

```python
np.random.seed(42)
# Generate synthetic data
```

```python
n_students = 200
study_hours = np.random.uniform(0, 10, n_students) 
attendance = np.random.uniform(0, 100, n_students)
```

```python
#attendance
```

```python
combined_score = 0.6 * study_hours + 0.4 * (attendance/10) 
```

```python
random_factor = np.random.normal(0, 1, n_students)
```

```python
random_factor
```

calculate with combined score with noise

```python
final_combined_score = combined_score + random_factor
```

```python
final_combined_score
```

determine pass / fail

```python
pass_fail = (final_combined_score > 5.0).astype(int)
```

```python
pass_fail
```

create dataframe

```python
students = pd.DataFrame({
    'Study Hours': study_hours,
    'Attendance': attendance,
    'Result': ["pass" if( x == 1) else "fail"  for x in pass_fail],
    "resultNumeric":pass_fail
})
```

```python
students
```

```python
students.head(10)
```

calculate the number of pass / fail

```python
print(f"passed student {np.sum(pass_fail)}")
print(f"failed student {len(pass_fail) - np.sum(pass_fail)}")
```

```python
sns.scatterplot(data=students, x='Attendance', y='Study Hours', hue='Result')
plt.show()
```

prepare the features and target

```python
x = students[['Study Hours', 'Attendance']].values
y = students['resultNumeric'].values
```

split data into training and testing 

```python
from sklearn.model_selection import train_test_split


X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=4, stratify=y)
```
