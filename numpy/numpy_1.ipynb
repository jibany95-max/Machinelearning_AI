```python
import numpy as np
```

```python
arr1 = np.array([[1,2,3,4,5]])
arr2 = np.array([[5,6,7,8,9]])
```

```python
arr3 = np.concatenate([arr1, arr2], axis=0)
arr3
```

```python
arr4 = np.concatenate([arr1, arr2], axis = 1)
arr4
```
