### converting categorical data into numerical

```python
df["Sex"] = df["Sex"].map({"male": 1, "female":0})
```

```python
df.head()
```

```python
df1 = df.copy()
```

```python
df1.head()
```

```python
df1 = pd.get_dummies(df1, columns=["Embarked"],dtype=int)
```

```python
df1.head()
```

```python
df1 = pd.get_dummies(df1, columns=["Embarked_C"],dtype= int, drop_first=True)
```

```python
df1.head()
```

```python
def fare_category(fare):
    if fare < 15:
        return "Low"
    elif fare < 40:
        return "Medium"
    else:
        return "High"
```

```python
df1["fare_category"] = df1["Fare"].apply(fare_category)
```

```python
df1.head()
```

### Remove the fare category column

```python
df1 = df1.drop(columns=["Fare"])
```

```python
df1.head()
```

```python
df1["Cabin"].values
```

### handling the null value in the cabin by filling 

```python
df1["Has_Cabin"] = np.where(df1["Cabin"].notna(), "Yes", "No")
```

```python
df1.head()
```

## Now remove the cabin from dataframe

```python
df1 = df1.drop(columns= ["Cabin"])
```

```python
df1["Has_Cabin"] = df1["Has_Cabin"].map({"Yes":1, "No":0})
```
