```python
dict2={
    "students":["student1","student2","student3","student4","student5","student6","student7"],
    "rollno":['005','006',np.nan,'005','011','012','013'],
    "gender":["MALE",np.nan,"FEMALE","FEMALE","male","female","female"],
    "MARK_OBTAINED":[65,66,34,56,20,np.nan,56]

}
```

```python
dict2
```

```python
student_df = pd.DataFrame(dict2)
```

```python
student_df
```

### To check  the null value

```python
student_df.isnull()
```

```python
student_df.isnull().sum()
```

```python
student_df.duplicated().sum()
```

## Remove the dupicate value

```python
student_df.drop_duplicates()
```

```python
student_df.dropna()
```

```python
gender_mode = student_df['gender'].mode()[0]

gender_mode
```

```python
marks_mean = student_df['MARK_OBTAINED'].mean()
marks_mean
```

```python
student_df['gender']=student_df['gender'].fillna(gender_mode)
student_df
```

```python
student_df['MARK_OBTAINED']=student_df['MARK_OBTAINED'].fillna(marks_mean)
```
