#### pie chart

```python
language = ["c++", "python", "java", "javascripts", "ruby"]
polularity = [56, 78, 34, 33, 19]
plt.pie(polularity, labels= language, autopct= '%.2f%%')
plt.show()
```

```python
language = ["c++", "python", "java", "javascripts", "ruby"]
polularity = [56, 78, 34, 33, 19]
explode = [0, 0.2 , 0.2, 0.3 , 0.2]
plt.pie(polularity, labels= language, autopct= '%.2f%%', explode= explode)

plt.show()
```

### showing multiple plot

```python
plt.figure(figsize=(12, 6))

plt.subplot(2, 2, 1)
language = ["c++", "python", "java", "javascripts", "ruby"]
polularity = [56, 78, 34, 33, 19]
plt.bar(language, polularity, color = 'green')
plt.title("Language and their poluarity")

plt.subplot(2, 2, 2)
x = np.random.normal(50, 2.5, 100)
plt.hist(x, bins = 20)
plt.title("Histogram")

plt.subplot(2, 2, 3)
language = ["c++", "python", "java", "javascripts", "ruby"]
polularity = [56, 78, 34, 33, 19]
plt.pie(polularity, labels= language, autopct= '%.2f%%')
plt.title("Pie chart ")

plt.subplot(2, 2, 4)
data = np.random.normal(50, 3, 100)
plt.boxplot(data)
plt.title("Box plot")
```
